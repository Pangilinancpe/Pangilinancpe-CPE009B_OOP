1. Read a text file from your computer.

```python
def read_file(filepath):
    with open(filepath, "r") as f:
        return f.read()

print(read_file("sample.txt"))
```

2. Read first n lines of a text file.


```python
def read_first_n_lines(filename, n):
    print(f"--- First {n} Lines ---")
    with open(filename, "r") as file:
        for i in range(n):
            line = file.readline()
            if not line:
                break
            print(line.strip())

read_first_n_lines("sample.txt", 3)
```

3. Append text to a file and display the text file's contents.

```python
def append_and_display(filename, text_to_append):
    with open(filename, "a") as file:
        file.write(text_to_append + "\n")

    print("--- Updated File Content ---")
    with open(filename, "r") as file:
        print(file.read())

append_and_display("sample.txt", "Let's build something fanstastic project to shock them.")
```

4. Ask the user for n-line and read last n lines of a file, display.

```python
def read_last_n_lines(filename):
    try:
        n = int(input("Enter the number of lines to read from the end: "))
        with open(filename, "r") as file:
            lines = file.readlines()
            last_n = lines[-n:] if n <= len(lines) else lines

            print(f"\n--- Last {n} Lines ---")
            for line in last_n:
                print(line.strip())
    except ValueError:
        print("Please enter a valid whole number.")

read_last_n_lines("sample.txt")
```

5. Read a file line by line and store it into a list. Display the list

```python
def read_into_list(filepath):
    with open(filepath, "r") as f:
        lines_list = [line.rstrip("\n") for line in f]
    print(lines_list)
    return lines_list

read_into_list("sample.txt")
```

6. Read a file line by line store it into a variable. Display the variable.

```python
def file_to_variable(filename):
    file_contents_var = ""
    with open(filename, "r") as file:
        for line in file:
            file_contents_var += line  # Chains each string into the variable

    print("--- File Stored inside a Variable ---")
    print(file_contents_var)

file_to_variable("sample.txt")
```

7. Find the longest words in a text file. Display the longest word.

```python
import string

def find_longest_word(filepath):
    longest = ""
    with open(filepath, "r") as f:
        for line in f:
            for word in line.split():
                cleaned = word.strip(string.punctuation)
                if len(cleaned) > len(longest):
                    longest = cleaned
    print(longest)
    return longest

find_longest_word("sample.txt")
```

8. Count the number of lines in a text file. Display the number of lines.


```python
def count_lines(filepath):
    with open(filepath, "r") as f:
        count = sum(1 for _ in f)
    print("--- Line Count ---")
    print(f"Total number of lines: {count}")
    return count

count_lines("sample.txt")
```

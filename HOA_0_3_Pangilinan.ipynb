```python
def add(x, y):
    # Return the sum of the two numbers
    return x + y


def subtract(x, y):
    # Return the difference (x minus y)
    return x - y


def multiply(x, y):
    # Return the product of the two numbers
    return x * y


def divide(x, y):
    if y == 0:
        raise ZeroDivisionError("Cannot divide by zero.")
    # Return the quotient (x divided by y)
    return x / y


def get_number(prompt):
    # Loop forever until the user enters something that converts to a float
    while True:
        try:
            # Attempt to read input and cast it to a float
            return float(input(prompt))
        except ValueError:
            # input() succeeded but float() failed -> not a number
            print("Invalid input! Please enter a valid number.")


def main():
    # Program header
    print("--- Calculator ---")

    # Collect both operands, validated by get_number()
    num1 = get_number("Enter the first number: ")
    num2 = get_number("Enter the second number: ")

    # Section header for the results block
    print("\n--- Results ---")

    # Sum outpu
    print(f"Sum:        {add(num1, num2)}")

    # Difference output
    print(f"Difference: {subtract(num1, num2)}")

    # Product output
    print(f"Product:    {multiply(num1, num2)}")

    # Quotient output and we use the try and except statement
    try:
        print(f"Quotient:   {divide(num1, num2)}")
    except ZeroDivisionError as e:
        # Catch the custom error raised inside divide() and display its message
        print(f"Quotient:   Error ({e})")


if __name__ == "__main__":
    main()
```

Conclusion

Based on what I've learned on this activity, we created a simple python calculator was developed to perform four basic arithmetic operations which are addition, subtraction, multiplication, and division, using dedicated functions for each. An exception handling was applied at two levels using the try/except loop around user input to catch non-numeric entries ValueError, and an inside the division function to catch and raise a custom ZeroDivisionError when the second operand is zero. Using this to demonstrates how input validation and error handling can prevent a program from invalid output, while keeping the arithmetic logic itself simple from the error-handling code.

```python
name = input("Enter your name: ") # Prompt the user to enter their name and store it in the 'name' variable
age = input("Enter your age: ") # enter their age and store it in the 'age' variable
gender = input("Enter your gender: ")  # enter their age and store it in the 'age' variable
address = input("Enter your address: ")  # enter their address and store it in the 'address' variable
contact = input("Enter your contact: ")  # enter their contact and store it in the 'contact' variable
email = input("Enter your email: ")  # enter their email and store it in the 'email' variable

with open("BPangilinan.txt", "w") as f:
    f.write(f"Name: {name}\n")  # Write the collected name into the file
    f.write(f"Age: {age}\n")  # Write the collected age into the file
    f.write(f"Gender: {gender}\n")  # Write the collected gender into the file
    f.write(f"Address: {address}\n")  # Write the collected address into the file
    f.write(f"Contact: {contact}\n")  # Write the collected contact into the file
    f.write(f"Email: {email}\n")  # Write the collected name into the file

print("Information saved to BPangilinan.txt") # confirmation message to the console indicating the file was saved

```

```python
def read_file(filename="BPangilinan.txt"): # Reads the contents of the file and returns it as a string.

    try: # Start a try block to handle potential errors (like the file missing)
        with open(filename, "r") as f: # The 'with' keyword ensures the file automatically closes when the block ends
            content = f.read() # # Read the entire contents of the file and store it in the 'content' variable
        return content # Return the saved file contents back to function was called
    except FileNotFoundError: # If the file does not exist, catch the FileNotFoundError exception
        return f"Error: The file '{filename}' does not exist yet." # Return a friendly error message
```

```python
# Prints the provided content to the console
def print_info(content):
    print("--- Stored Information ---")
    print(content)
    print("--------------------------")
```

```python
# Call the read function and store the result
file_contents = read_file()

# Pass the result into the print function
print_info(file_contents)
```
